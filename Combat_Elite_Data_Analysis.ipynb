# **Player Segmentation through Cluster Analysis**

```python
# Preparing data for segmentation

# Total real money spent per player
total_iap_spend = data_in_app_purchases.groupby('user_id')['dollar_purchase_value'].sum()

# Total virtual money spent per player
total_virtual_spend = data_virtual_purchases.groupby('user_id')['gold_spend'].sum()

# Average finish position and total matches played per player
average_finish = data_matches.groupby('user_id')['finish_position'].mean()
total_matches = data_matches.groupby('user_id')['n_matches'].sum()

# Merge these metrics into a single DataFrame for clustering
player_metrics = pd.DataFrame({
    'total_iap_spend': total_iap_spend,
    'total_virtual_spend': total_virtual_spend,
    'average_finish': average_finish,
    'total_matches': total_matches
})

# Fill missing values (for players with no purchases or matches recorded in certain categories)
player_metrics.fillna(0, inplace=True)

# Overview of the prepared data
player_metrics.head()



from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Standardizing the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(player_metrics)

# Applying K-means clustering
# We choose a preliminary number of clusters; typically, this might need more fine-tuning or methods like the elbow method to find the optimal number.
kmeans = KMeans(n_clusters=4, random_state=0)
clusters = kmeans.fit_predict(scaled_features)

# Adding the cluster labels to the player metrics DataFrame
player_metrics['cluster'] = clusters

# Analyzing the characteristics of each cluster
cluster_summary = player_metrics.groupby('cluster').agg({
    'total_iap_spend': ['mean', 'median'],
    'total_virtual_spend': ['mean', 'median'],
    'average_finish': ['mean', 'median'],
    'total_matches': ['mean', 'median', 'count']
})

cluster_summary
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

# Prepare data for the decision tree
# Extract relevant features from in-app purchases and daily activity
iap_data = data_in_app_purchases[['user_id', 'activity_date', 'product_group']]
activity_data = data_daily_activity[['user_id', 'platform', 'activity_date']]

# Merge datasets on user_id and activity_date to get features from daily activity
merged_data = pd.merge(iap_data, activity_data, on=['user_id', 'activity_date'], how='left')

# Fill any missing values
merged_data.fillna('Unknown', inplace=True)

# Encode categorical variables
le_product_group = LabelEncoder()
le_platform = LabelEncoder()

merged_data['product_group_encoded'] = le_product_group.fit_transform(merged_data['product_group'])
merged_data['platform_encoded'] = le_platform.fit_transform(merged_data['platform'])

# Select features and target
features = merged_data[['platform_encoded']]
target = merged_data['product_group_encoded']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Train the decision tree classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = dt_classifier.predict(X_test)

# Evaluate the model
report = classification_report(y_test, y_pred, target_names=le_product_group.classes_)
report
```

Null Hypothesis (H0): There is no significant difference in the average in-app purchase spending between users on different platforms (iOS vs. Android).

Alternative Hypothesis (H1): There is a significant difference in the average in-app purchase spending between users on different platforms.**

```python
from scipy.stats import ttest_ind

# Extract relevant data
iap_data_platform = pd.merge(data_in_app_purchases, data_daily_activity[['user_id', 'platform']], on='user_id', how='left')

# Group by platform and calculate average spending
platform_spending = iap_data_platform.groupby('platform')['dollar_purchase_value'].mean()

# Separate data for t-test
ios_spending = iap_data_platform[iap_data_platform['platform'].str.lower() == 'ios']['dollar_purchase_value']
android_spending = iap_data_platform[iap_data_platform['platform'].str.lower() == 'android']['dollar_purchase_value']

# Perform t-test
t_stat, p_value = ttest_ind(ios_spending, android_spending, equal_var=False)  # Use Welch's t-test for unequal variances

platform_spending, t_stat, p_value
```

Hypothesis 4: Relationship Between Engagement and Virtual Currency Spending
Null Hypothesis (H0): There is no significant relationship between the frequency of user engagement (daily activity) and the amount of virtual currency spent.

Alternative Hypothesis (H1): There is a significant relationship between the frequency of user engagement (daily activity) and the amount of virtual currency spent.

```python
from scipy.stats import pearsonr


# Calculate the frequency of user engagement (total number of active days)
user_engagement = data_daily_activity.groupby('user_id').size().reset_index(name='active_days')

# Aggregate the total virtual currency spent by each user
total_virtual_spending = data_virtual_purchases.groupby('user_id')['gold_spend'].sum().reset_index()

# Merge the datasets to align user engagement with spending
engagement_spending_data = pd.merge(user_engagement, total_virtual_spending, on='user_id', how='inner')

# Perform Pearson correlation test
correlation_coefficient, p_value = np.corrcoef(engagement_spending_data['active_days'], engagement_spending_data['gold_spend'])[0, 1], pearsonr(engagement_spending_data['active_days'], engagement_spending_data['gold_spend'])[1]

correlation_coefficient, p_value
```

Hypothesis 5: Effect of Acquisition Date on Long-Term Spending
Null Hypothesis (H0): The acquisition date (when a user first started playing) does not significantly affect the long-term spending behavior of users.

Alternative Hypothesis (H1): The acquisition date significantly affects the long-term spending behavior of users.


```python
from datetime import datetime
import statsmodels.api as sm

from datetime import datetime
import statsmodels.api as sm

# Extract the first activity date (acquisition date) for each user
acquisition_dates = data_daily_activity.groupby('user_id')['activity_date'].min().reset_index()
acquisition_dates.rename(columns={'activity_date': 'acquisition_date'}, inplace=True)

# Convert acquisition_date to datetime
acquisition_dates['acquisition_date'] = pd.to_datetime(acquisition_dates['acquisition_date'])

# Calculate the total long-term spending for each user (sum of both in-app and virtual purchases)
iap_spending = data_in_app_purchases.groupby('user_id')['dollar_purchase_value'].sum().reset_index()
virtual_spending = data_virtual_purchases.groupby('user_id')['gold_spend'].sum().reset_index()

# Merge spending data
total_spending = pd.merge(iap_spending, virtual_spending, on='user_id', how='outer').fillna(0)
total_spending['total_spending'] = total_spending['dollar_purchase_value'] + total_spending['gold_spend']

# Merge acquisition dates with spending data
acquisition_spending_data = pd.merge(acquisition_dates, total_spending[['user_id', 'total_spending']], on='user_id', how='inner')

# Convert acquisition_date to ordinal for regression analysis
acquisition_spending_data['acquisition_date_ordinal'] = acquisition_spending_data['acquisition_date'].apply(lambda x: x.toordinal())

# Perform regression analysis
X = acquisition_spending_data['acquisition_date_ordinal']
y = acquisition_spending_data['total_spending']
X = sm.add_constant(X)  # Adds a constant term to the predictor

model = sm.OLS(y, X).fit()
summary = model.summary()

summary

```
